## 3.3 Handler、Thread和HandlerThread 的区别
* Handler：异步消息处理机制
* Thread：Java 的多线程类，是Java进程中最小执行运算单位，用于给子类继承
* HandlerThread：一个继承自Thread的Handler类，本质是个线程类，内部创建了Looper对象，可以进行消息循环

## 3.4 子线程如何使用Handler？

## 3.5 为什么在子线程中创建Handler，会抛异常？
不能在没有调用Looper.prepare() 方法时创建Handler，否则会抛出异常。
Handler 的构造函数中，会进行判断，如果当前线程没有Looper，就会抛出异常。
